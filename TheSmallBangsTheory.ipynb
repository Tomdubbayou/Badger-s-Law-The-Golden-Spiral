```python
# The Small Bangs Simulation
# By Tom & Loo, 2025

import numpy as np
import matplotlib.pyplot as plt

# Constants and spiral growth
steps = 1500
phi = (1 + np.sqrt(5)) / 2  # Golden ratio
theta = np.linspace(0, 10 * np.pi, steps)
r = 0.05 * theta

# Simulate harmonic pressure accumulation
pressure = np.sin(2.5 * theta) + 0.4 * np.cos(4.2 * theta) + 0.2 * np.sin(phi * theta)
threshold = 0.95
burst_indices = np.where(pressure > threshold)[0]

# Cartesian spiral for burst visualization
x = r * np.cos(theta)
y = r * np.sin(theta)

# Spiral Plot with Burst Points
plt.figure(figsize=(8, 8))
plt.plot(x, y, lw=0.8, label='Spiral Growth')
plt.scatter(x[burst_indices], y[burst_indices], color='crimson', s=30, label='Small Bangs')
plt.title('The Small Bangs: Spiral Harmonic Burst Map')
plt.axis('equal')
plt.legend()
plt.grid(True)
plt.show()

# Pressure Plot
plt.figure(figsize=(10, 4))
plt.plot(theta, pressure, label='Harmonic Pressure')
plt.axhline(threshold, color='red', linestyle='--', label='Burst Threshold')
plt.scatter(theta[burst_indices], pressure[burst_indices], color='red', label='Small Bang Events')
plt.title('Harmonic Pressure Across Spiral Turns')
plt.xlabel('Theta (spiral angle)')
plt.ylabel('Pressure')
plt.legend()
plt.grid(True)
plt.show()
```
